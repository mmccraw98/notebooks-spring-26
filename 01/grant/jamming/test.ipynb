```python
import jax.numpy as jnp
import jax
import jaxdem as jd
jax.config.update("jax_enable_x64", True)
import numpy as np
import os
from bump_utils import render
```

```python
data_root = '/home/mmccraw/dev/data/26-01-01/grant/jamming/3d'
for name in os.listdir(data_root):
    root = os.path.join(data_root, name)
    if not os.path.exists(os.path.join(root, 'state.h5')):
        continue
    state = jd.utils.h5.load(os.path.join(root, 'state.h5'))
    system = jd.utils.h5.load(os.path.join(root, 'system.h5'))
    render(state, system, f'{name}.png')
```
